### ***Missing Values* - Valores NaN presentes no dataset**
Algumas propriedades de elementos de número atômico maiores não possuem registros, porque a meia-vida dos elementos criados artificialmente é muito pequena e impossibilita as medições apropriadas.


```python
df.describe().agg(["count", "mean", "std", "median", "min", "max"])
```

Todas as propriedades estão compreendidas no intervalo de números maiores que zero. Isso é importante porque evidencia que as temperaturas estão medidas em Kelvin (K).

```python
# Troca dos valores NaN por um número próximo de 0, que no projeto irá significar a existência de uma coluna de altura mínima.
df = df.fillna(0.0001)
df[df["Symbol"] == "Fr"]
```

```python
# Seleciona todas as colunas a partir do índice 2
column_names = df.iloc[:, 2:].columns

# Constroi a figura e os subplots com tamanho e dpi definidos
fig, ax = plt.subplots(len(column_names), figsize = (20, 40), dpi = 70)

# Tabela de cores

colors = ["r", "y", "grey", "blue", "cyan", "green", "purple"]

# Laço pra construção de cada tabela
for i, column in enumerate(column_names):
  ax[i].bar(df[column].index, df[column], color = colors[i], edgecolor = 'black')
  ax[i].set_title(column)

plt.show()
```

Nos gráficos acima, é possível perceber que cada para cada propriedade, o dado se comporta de maneira diferente. Alguns crescem de forma linear conforme o número atômico (Z) também aumenta, como é o caso da massa atômica. Outros possuem padrões visíveis, mas que também se alteram devido a forma peculiar da tabela, como o caso do Raio Atômico.

# **2. Modelo matemático**

A regra utilizada para a construção das colunas da tabela periódica em 3D sugere que cada coluna siga uma conversão (ou *mapping*) de valores compreendidos na série da cada coluna baseados no seu amplitude (ou *range*, da estatística) do seu intervalo. A expressão que sugere cada altura é:

$$
H = h_{max} \cdot \dfrac{x_i}{x_{max} - x_{min}}
$$

Onde:

**H**: Altura efetiva da coluna para cada elemento;

**$h_{máx}$**: Altura máxima definida (parâmetro construtivo);

**$x_i$**: Valor observado da propriedade para um elemento i qualquer;

**$x_{max}$**: Valor máximo da série para a propriedade observada;

**$x_{min}$**: Valor mínimo da série para a propriedade observada.

***

Para essa análise, consideraremos que o valor máximo que uma coluna pode ter é de **50 milímetros**. 

```python
df.head()
```

```python
# Aplicando a função que retorna a altura (_height) de cada coluna, temos:

h_max = 50

df_heights = df.apply(lambda x: round(h_max * (x)/(x.max() - x.min()), 4) if x.name in ['AtomicMass', 
                                                                                        'AtomicRadius', 
                                                                                        'Electronegativity', 
                                                                                        'Density', 
                                                                                        'MeltingPoint', 
                                                                                        'BoilingPoint', 
                                                                                        'SpecificHeat'] else x)
```

```python
df_heights
```

```python
# Visualizando novamente os gráficos que foram convertidos

# Seleciona todas as colunas a partir do índice 2
column_names = df_heights.iloc[:, 2:].columns

# Constroi a figura e os subplots com tamanho e dpi definidos
fig, ax = plt.subplots(len(column_names), figsize = (20, 40), dpi = 70)

# Tabela de cores
colors = ["r", "y", "grey", "blue", "cyan", "green", "purple"]

# Laço pra construção de cada tabela
for i, column in enumerate(column_names):
  ax[i].bar(df_heights[column].index, df_heights[column], color = colors[i], edgecolor = 'black')
  ax[i].set_title(column)

plt.show()
```

Os gráficos devem manter o mesmo aspecto, significando apenas que foram convertidos em uma proporção para caber em uma escala de 0 a 50mm.

# **3. Integração com Fusion 360**

A integração com o software de modelagem 3D - Autodesk Fusion 360 - foi feita através do uso de um add-in (uma instalação que atua como um upgrade) chamado [*Parameter I/O*](https://apps.autodesk.com/FUSION/pt/Detail/Index?id=1801418194626000805&appLang=en&os=Win64). Nele é possível fazer uma leitura de um CSV contendo todos os parâmetros referentes a cada elemento da tabela periódica, que fora montada em uma espécie de modelo 3D base para que sejam inseridas as alturas.

A formatação do CSV que deve ser utilizada é:

> nome_do_parametro, mm, medida + 'mm', comentario

> Ex: hidrogenio, mm, 40.23 mm, 



```python
df_heights.head(3)
```
